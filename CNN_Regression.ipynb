```python
import keras
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from keras.models import Sequential
from keras.layers import Dense
```

```python
dataset = pd.read_csv("Dataset2_correct.csv")
signal_data = dataset.drop(columns = ["max_param", "amp_threshold", "x_position"])
target_data = dataset[["x_position"]]
train_signals, test_signals, train_targets, test_targets = train_test_split(signal_data, target_data, test_size = 0.3)
train_signals, train_targets, test_signals, test_targets = pd.DataFrame(train_signals).to_numpy(), pd.DataFrame(train_targets).to_numpy(), pd.DataFrame(test_signals).to_numpy(), pd.DataFrame(test_targets).to_numpy()
train_signals = np.reshape(train_signals, (train_signals.shape[0], train_signals.shape[1], 1))
train_targets = np.reshape(train_targets, (train_targets.shape[0], train_targets.shape[1], 1))
test_targets = np.reshape(test_targets, (test_targets.shape[0], test_targets.shape[1], 1))
test_signals = np.reshape(test_signals, (test_signals.shape[0], test_signals.shape[1], 1))
print(train_signals.shape, test_signals.shape, train_targets.shape, test_targets.shape)
```

```python
model = tf.keras.Sequential()
model.add(layers.Conv1D(filters=32, kernel_size=3, activation='relu', input_shape = train_signals.shape[1:]))
model.add(layers.MaxPooling1D(pool_size=3))

#model.add(layers.Conv1D(filters=64, kernel_size=3, activation='relu'))
#model.add(layers.MaxPooling1D(pool_size=2))

model.add(layers.Flatten())

# Add fully connected layers
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(1))
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(train_signals, train_targets, epochs=400, batch_size=32, validation_data=(test_signals, test_targets))
loss = model.evaluate(test_signals, test_targets)
print('Test Loss:', loss)
```

```python
prediction = (model.predict(test_signals))
print(prediction, test_targets)
```

```python
model.save("Threshold_Param_Prediction.h5")
```
